```python
from typing import Sequence

def get_f1(pred: Sequence[float], gold: Sequence[float]):

    TP = 0
    TN = 0
    FP = 0
    FN = 0

    # 
    for sys_output, gold_standard in zip(pred, gold):
        if sys_output == 1 and gold_standard == 1:
            TP += 1
        elif sys_output == 0 and gold_standard == 0:
            TN += 1
        elif sys_output == 1 and gold_standard == 0:
            FP += 1
        elif sys_output == 0 and gold_standard == 1:
            FN += 1

        


    # calculate precision and recall
    precision = TP / (TP + FP) if (TP + FP) > 0 else 0
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0

    # calculate F1_socre
    f1 = (2 * precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    return f1


# read input prediictions
pred = [1, 0, 1, 0, 0, 1, 0, 1, 1, 1, 1, 0, 0, 1]  
gold = [1,0,1,1,0,1,0,1,1,1,1,0,0,1,]

f1 = get_f1(pred, gold)

print(f"{f1:.2f}")
```

## Binary classification (Binominal logistic regression)

- it often use `sigmoid function` = Converting any number into a probability between 0 - 1
- it often sets the threshold 0.5
- $$\sigma(z) = \frac{1}{1+e^{-z}}$$

## Multi-class classification (Multinomial logistic regression)
- `Softmax Function` is often used
- Convert multiple raw scores into probabilities that sum to 1
- $$P(y = k) = \frac{e^{z_k}}{\sum_{j=1}^{K} e^{z_j}}$$
